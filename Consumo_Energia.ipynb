```python
import pandas as pd
import numpy as np


df= pd.read_csv('Fuentes_Consumo_Energia.csv')
```

```python
# elimino la columna llamada Code de forma permanente por usar "inplace=True". Axis se usa para indicar el nombre de las columnas
df.drop('Code', axis=1, inplace=True)
```

```python
#Elimina las filas en donde se encuentran valores faltantes en todo el data frame de manera permanente. El thresh le indica la cantidad de valores que deben ser no nulos
df.dropna(inplace=True, thresh=9)
```

```python
df.sort_values('Year') #ordenar los datos por año
df.tail() #mostrar el final del dataframe
```

Convertir los consumos energéticos registrados en ExaJulios a Teravatio/hora para poder calcular, dentro de la Entidad 'Mundo', el mayor consumo del año 2019

```python

df1= pd.read_csv('Fuentes_Consumo_Energia.csv', delimiter = ',',encoding = "utf-8")
df1['Coal_Consumption_EJ-de-TWh']=df1['Coal_Consumption_EJ'].apply(lambda x:x * 277.778)
df1['Gas_Consumption_EJ-de-TWh']=df1['Gas_Consumption_EJ'].apply(lambda x:x * 277.778)
df1['Oil_Consumption_EJ-de-TWh']=df1['Oil_Consumption_EJ'].apply(lambda x:x * 277.778)
df1['consumo_total']=df1.apply( lambda row: row['Coal_Consumption_EJ-de-TWh']+    row['Gas_Consumption_EJ-de-TWh']+    row['Oil_Consumption_EJ-de-TWh']+    row['Geo_Biomass_Other_TWh']+    row['Hydro_Generation_TWh']+    row['Nuclear_Generation_TWh']+    row['Solar_Generation_TWh']+    row['Wind_Generation_TWh'], axis=1)
valorfinal=round(df1['consumo_total'][(df1.Entity=='World')&(df1.Year==2019)].sum(),2)
valorfinal
```

```python
#Reiniciar el index sin conservar en una nueva columna el anterior. Para eso se usa drop=True
df.reset_index(inplace=True,drop=True)
```

### Trabajando con una sola fila

```python
#buscar en la columna entity el valor world y ademas en la columna Year el valor 2019
df[(df['Entity']=='World')&(df['Year']==2019)]
```

```python
#Para hacer un Slicing usamos iloc [filas o fila_inicio:fila_final, columna o columna_inicio:columna_final] Teniendo en cuenta que el valor final no se toma. Si pasamos un solo valor se toma tal cual.

#Esto sería trabajando con una sola fila
suma_de_twh = df.iloc[4138,5:10].sum() #suma el valor contenido en la fila 4138 y las columnas 5, 6, 7, 8 y 9
convertidos = suma_de_twh/277.778 
suma_columna3y4y10 = df.iloc[4138,3:5].sum()+df.iloc[4138,10] #suma los valores contenidos en la fila 4138 y las columnas 3, 4 y 10
suma_total = convertidos+suma_columna3y4y10
```

```python
print(round(suma_total,2))
```
